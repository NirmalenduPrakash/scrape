```python
import scrapy
import time
from googleapiclient.discovery import build
import requests
import json
```

```python
class Course:
    def __init__(self,inputFile):
        self.courses=open(inputFile, "r").readlines()
        self.urls=[]
    def getUrls(self):    
        for course in self.courses:
            #url=self.google_search(course,my_api_key,my_cse_id,num=5)
            url=self.bing_search(course,customConfigId,subscriptionKey)
            time.sleep(2)
            self.urls.append(url)
        print(self.urls)    
        with open('urls.txt', 'w') as f:
            for item in self.urls:
                f.write("%s\n" % item)       
    def google_search(self,search_term, api_key, cse_id, **kwargs):
        service = build("customsearch", "v1", developerKey=api_key,cache_discovery=False)
        res = service.cse().list(q=search_term, cx=cse_id, **kwargs).execute()
        if('items' in res.keys()):
            return res['items'][0]['link']
        return ''
    def bing_search(self,searchTerm,customConfigId,subscriptionKey):
        try:
            url = 'https://api.cognitive.microsoft.com/bingcustomsearch/v7.0/search?' + 'q=' + searchTerm + '&' + 'customconfig=' + customConfigId
            r = requests.get(url, headers={'Ocp-Apim-Subscription-Key': subscriptionKey})
            return r.json()["webPages"]["value"][0]['url']
        except Exception:
            print(Exception)
            return 'Not Available'
```

```python
class ScrapeUtil(scrapy.Spider):
    name = 'MySpider'

    def __init__(self, dataFile):
        #Course(dataFile).getUrls()
        #time.sleep(5)
        self.urls = open('paths.txt', "r").readlines()
    def parse(self, response):
        XPATH_Overview= '//div[contains(@id,"overview")]'
        XPATH_Lecturer = '//div[contains(@id,"instructors")]/div[contains(@class,"row")]/div[1]/h3/text()'

        REFERENCE_No=response.xpath(XPATH_Overview+'/table/tbody/tr[0]/td/text()').extract()    
        PART_OF=response.xpath(XPATH_Overview+'/table/tbody/tr[1]/td/text()').extract() 
        DURATION=response.xpath(XPATH_Overview+'/table/tbody/tr[2]/td/text()').extract()
        COURSE_TIME=response.xpath(XPATH_Overview+'/table/tbody/tr[3]/td/text()').extract()  
        ENQUIRY=response.xpath(XPATH_Overview+'/table/tbody/tr[4]/td/text()').extract()

        INSTRUCTORS=response.xpath(XPATH_Lecturer).extract()    
        
        yield {'Reference No':REFERENCE_No,'Part of':PART_OF,'Duration':DURATION,
        'Course time':COURSE_TIME,'Enquiry':ENQUIRY,'Instructors':[instructor for instructor in INSTRUCTORS]}

    def start_requests(self):
        # headers = {
        #     'accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,*/*;q=0.8',
        #     'cache-control': 'no-cache',
        #     'User-Agent': 'Mozilla/5.0'
        # }
        for url in self.urls:
            yield scrapy.Request(url, self.parse,headers=headers)
```
